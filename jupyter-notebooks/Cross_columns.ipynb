### Active Power (P)
Active power, measured in watts (W), is the actual power that can be used to perform work, such as turning a motor or lighting a bulb. In the context of a wind turbine, active power is the power produced by the turbine that is converted into electrical energy and can be used by consumers or fed into the electrical grid. It is the product of the voltage, current, and the cosine of the phase angle (power factor) between them in an AC (Alternating Current) system. Active power is what drives devices and charges batteries.

### Reactive Power (Q)
Reactive power, measured in volt-amperes reactive (VAR), is power that oscillates between the source and the load, without being consumed or used for work. It is required to maintain the voltage levels necessary for electrical systems to function properly. In AC systems, reactive power is generated by inductors and capacitors, which store and release energy with each cycle of the current. This power is necessary for creating magnetic fields in motors and for other functions that require a change in the energy storage but doesn't contribute to actual work. In the case of wind turbines, reactive power is important for maintaining voltage stability and power quality in the electrical grid.
